## Loading packages

```r
library("stringr")
library("phytools")
library("ggtree")
library("ggplot2")
library("ggpmisc")
library("deeptime")
```

## Loading data

```r
trait_model<-readRDS("ancestral_state.rds")
phy<-read.tree("Input/tree_orecto.tree")
```

### Preparing the tree

```r
phy$tip.label <- gsub("_", " ", phy$tip.label)
```

```r
node_states_biogeo <- trait_model[(length(phy$tip.label) + 1):(2*length(phy$tip.label) - 1),]
tip_states_biogeo <- trait_model[1:length(phy$tip.label),]
```

```r
color_list <- c("grey", "red", "blue", "yellow", "purple", "green", "cyan", "brown", "coral","chocolate", 
                "cornflowerblue", "deeppink", "darkorchid", "darkseagreen", "deepskyblue", "darkcyan", "darkgoldenrod")
```

```r
color_list <- setNames(color_list[1:length(unique(colnames(trait_model[,-18])))],sort(unique(colnames(trait_model[,-18]))))
```

```r
pies <- nodepie(node_states_biogeo, cols=1:17, color=color_list, alpha=1)
df<-tibble::tibble(node=as.numeric(node_states_biogeo$node), pies=pies)
```

```r
phylo_discrete <- ggtree(phy) +
           geom_tiplab(offset = 2, fontface = "italic", size = 2) +
           theme_bw() +
           theme(panel.border = element_blank(),
           legend.key = element_blank(),
           axis.ticks = element_blank(),
           axis.text.y = element_blank(),
           axis.text.x = element_blank(),
           panel.grid = element_blank(),
           panel.grid.minor = element_blank(), 
           panel.grid.major = element_blank(),
           panel.background = element_blank(),
           plot.background = element_rect(fill = "transparent",colour = NA)) + 
coord_geo(neg = T, pos = as.list(rep("bottom", 2)),
dat = list("epochs", "periods"), height = list(unit(1, "lines"), unit(1, "line")), size = list(2, 3), abbrv = "auto", skip = c("Pliocene", "Pleistocene", "Holocene"))
```

```r
phylo_discrete <- revts(phylo_discrete) +
           xlim(-250, 100) + ylim(0, 40)
```

```r
phylo_discrete <- revts(phylo_discrete) +
           xlim(-250, 100) + ylim(0, 40)
```

```r
phy_node <- phylo_discrete %<+% df
```

```r
phy_complete <- phy_node + geom_plot(data=td_filter(!isTip), mapping=aes(x=x,y=y, label=pies), vp.width=0.04, vp.height=0.04, hjust=0.6, vjust=0.6)
```

```r
vec <- c()
for(i in 1:nrow(tip_states_biogeo)){
    temp_tip_states_biogeo <- tip_states_biogeo[,-ncol(tip_states_biogeo)]
    vec <- c(vec, colnames(temp_tip_states_biogeo)[which(temp_tip_states_biogeo[i,]==1)])
}
```

```r
vec <- as.factor(vec)
```

```r
levels(vec) <- c(levels(vec), setdiff(rownames(tip_states_biogeo[-1]), levels(vec)))
```

```r
df_temp <- data.frame(Species = phy$tip.label, vec)
```

```r
phy_comp_2_0 <- phy_complete %<+% df_temp
```

```r
phy_comp_2_0 + geom_tippoint(aes(color = as.character(vec)), show.legend = TRUE) + 
    scale_color_manual(values = color_list, drop = TRUE, limits = names(color_list)) +
    theme(legend.position = c(0.15, 0.875)) + 
guides(color = guide_legend(ncol=4, title = "Biogeographic regions")) 
```
